## Explicação da hipótese e mecanismo

### A OPEP

A OPEP (Organização dos Países Exportadores de Petróleo), criada em 1961, funciona como uma espécie de cartel que unifica as leis e preços dos barris dos principais exportadores de petróleo, buscando, assim, estabilizar o mercado de petróleo. Ou seja, os paises participantes da OPEP conseguem agregar mais valor aos seus barris e, em teoria, conseguem ganhar a mesma quantidade de dinheiro produzindo menos.

Para saber mais, acesse o site da OPEP: https://www.opec.org/opec_web/en/about_us/23.htm

Os países que já participaram da OPEP com seus anos de entrada e, se houver saída, são:

### Mecanismo

Sabendo da premissa do funcionamento da OPEP, na teoria a organização buscaria agregar mais valor do que o normal aos barris de petróleo exportados pelos países pertencentes ao grupo. Porém, como mostrado no artigo abaixo, a Arábia Saudita apresenta um papel central e dominante na OPEP, caracterizando, na prática, a organização como um grupo com objetivos heterogêneos. Além disso, o artigo também analisa o sucesso da OPEP como organização à longo prazo e conclui que "apesar do sucesso de intervenções pontuais no mercado em momentos de crise e do impacto que suas decisões exercem sobre o preço do petróleo, a OPEP não é capaz de unir-se em torno de uma estratégia de longo prazo e, por isso, não dispõe credibilidade para tornar sua meta oficial de preços um bom parâmetro de mercado".

Sendo assim, espera-se que a produção de petróleo em cada país não seja modificada devido a participação do mesmo na OPEP, já que esta não trata-se da produção de petróleo, mas do preço dos barris exportados. Porém, espera-se nesse trabalho que o PIB desses países não cresca muito mais rapidamente do que o PIB dos países exportadores de petróleo que não participam da OPEP, pois, como observado pela pesquisa, devido aos interesses distintos dos participantes do grupo, a OPEP não consegue criar uma estratégia de mercado a longo prazo, falhando, assim, com o crescimento ao longo prazo do PIB.

### Ao entrar na OPEP um país aumenta consideravelmente sua produção de petróleo?

Para analizar essa pergunta foi escolhido observar os dados de 3 países (Gabão, Equador e Nigéria) ao longos dos anos de 1965 até 2000. A escolha dos países foi feita devido aos anos de entrada na OPEP, já que muitos entraram muito próximos, ou até antes, de 1965, ou então, entraram muito recentemente, o que prejudica a análise desejada.

```python
#Recortando a tabela pegando apenas os países que desejamos utilizar
GNE = produtores_petroleo_idx.loc[["Gabon", "Nigeria", "Ecuador" ],:]

#Recortar os anos que desejados
P3 = GNE[GNE.columns[0:36]]

#Transpondo a tabela 
GNE_transpose = P3.transpose()
```

```python
#Colocando parâmetros para melhor vizualização dos gráficos 
plt.rcParams["figure.figsize"] = [25,13]
plt.rcParams["axes.titlesize"] = 29
plt.rcParams["xtick.labelsize"] = 17
plt.rcParams["ytick.labelsize"] = 17
plt.rcParams["axes.labelsize"] = 20
plt.rcParams["lines.markersize"] = 18

#Plotando o gráfico
GNE_transpose.plot(kind='bar', color=('yellow', 'salmon', 'green'))
plt.ylabel('Produção de Petrólro')
plt.xlabel('Anos')
plt.title('Produção de petróleo de 1965 até 2005')
```

Pode-se pressupor que para o Gabão, que entrou na OPEP em 1975 e saiu em 1994, participar do cartel não alterou a produção interna de petróleo. O país já passava por um aumento de produção e depois que entrou no cartel chegou ao seu pico diminuindo logo depois. Pode-se ver também que, mesmo deixando o cartel em 1994, a produção depetróleo não é alterada. 

Para a Nigéria, que ingressou em 1971, pode-se notar um alto crescimento na produção nos anos de 1972, 1973 e 1974. Porém, ao fazer a analize do gráfico, nota-se que o país já apresentava um movimento alto de crescimento antes mesmo de entrar na OPEP. Sendo assim, não pode-se chegar à uma conclusão se esse aumento foi ou não impulsionado pela participação na OPEP ou não. 

Em 1973 o Equador inicia a atuar como um membro da organização. Pode-se notar um altíssimo aumento na prdução de petróleo nesse ano, que se manteve ao longo dos outros anos. O que talvez possa indicar que ingressar na OPEP o Equador buscou aumentar sua produção anual de petróleo. Porém, ao sair do cartel, em 1992, a produção continua a crescer, mesmo que em baixas taxas.

### Análise do PIB dos países

Para essa análise se unirão os dados coletados do crescimento PIB dos países com os dados coletados da produção de petróleo anual e se buscará vizualizar se o PIB dos países pertencentes à OPEP cresce mais do que o PIB dos outros países produtores de petróleo

```python
#Recortando as tabelas com apenas os anos que desejamos utilizar
PIB_1975 = pib_idx.loc[:,[1975, "OPEP"]].dropna()
PIB_1997 = pib_idx2.loc[:,[1997, "OPEP"]].dropna()
PIB_2013 = pib_idx3.loc[:,[2013, "OPEP"]].dropna()
```

```python
#Fazendo a junção de tabelas
PIB_PP_1975 = OPEP_1975.join(PIB_1975, how='inner', lsuffix='_petroleo', rsuffix='_PIB')
PIB_PP_1997 = OPEP_1997.join(PIB_1997, how='inner', lsuffix='_petroleo', rsuffix='_PIB')
PIB_PP_2013 = OPEP_2013.join(PIB_2013, how='inner', lsuffix='_petroleo', rsuffix='_PIB')
```

```python
#Definindo a função para fazermos uma reta de tendência nos gráficos que serão plotados
def reta(dados_x, dados_y, c):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()

    print('Ajuste linear entre X3 e X4:')
    print('Coeficiente linear:', b)
    print('Coeficiente angular:', a)
    plt.title('x versus y')
    plt.xlabel('eixo x')
    plt.ylabel('eixo y')

    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color=c)

    plt.tight_layout()
```

```python
#Colocando parâmetros para melhor vizualização dos gráficos
plt.rcParams["figure.figsize"] = [16,9]
plt.rcParams["axes.titlesize"] = 25
plt.rcParams["xtick.labelsize"] = 15
plt.rcParams["ytick.labelsize"] = 15
plt.rcParams["axes.labelsize"] = 17
plt.rcParams["lines.markersize"] = 15

#Plotando o gráfico com os países pertencentes à OPEP em 1975 e colocando uma reta de tendência 
plt.scatter(PIB_PP_1975.loc[PIB_PP_1975.OPEP_petroleo == "OPEP","1975_petroleo"], PIB_PP_1975.loc[PIB_PP_1975.OPEP_PIB == "OPEP", "1975_PIB"], color="turquoise", label = "Países OPEP")
reta(PIB_PP_1975.loc[PIB_PP_1975.OPEP_petroleo == "OPEP","1975_petroleo"], PIB_PP_1975.loc[PIB_PP_1975.OPEP_PIB == "OPEP", "1975_PIB"], "turquoise", )

#Plotando o gráfico com os países não pertencentes à OPEP em 1975 e colocando uma reta de tendência 
plt.scatter(PIB_PP_1975.loc[PIB_PP_1975.OPEP_petroleo == "Não OPEP","1975_petroleo"], PIB_PP_1975.loc[PIB_PP_1975.OPEP_PIB == "Não OPEP", "1975_PIB"], color="indigo", label = "Não OPEP")
reta(PIB_PP_1975.loc[PIB_PP_1975.OPEP_petroleo == "Não OPEP","1975_petroleo"], PIB_PP_1975.loc[PIB_PP_1975.OPEP_PIB == "Não OPEP", "1975_PIB"], "indigo")

plt.grid()
plt.title('Produção de petróleo x Crescimento anual do PIB em 1975')
plt.ylabel('Crescimento anual do PIB')
plt.xlabel('Produção de petróleo')
plt.legend()
```

```python
#Plotando o gráfico com os países pertencentes à OPEP em 1997 e colocando uma reta de tendência 
plt.scatter(PIB_PP_1997.loc[PIB_PP_1997.OPEP_petroleo == "OPEP","1997_petroleo"], PIB_PP_1997.loc[PIB_PP_1997.OPEP_PIB == "OPEP", "1997_PIB"], color="turquoise", label = "Países OPEP")
reta(PIB_PP_1997.loc[PIB_PP_1997.OPEP_petroleo == "OPEP","1997_petroleo"], PIB_PP_1997.loc[PIB_PP_1997.OPEP_PIB == "OPEP", "1997_PIB"], "turquoise")

#Plotando o gráfico com os países não pertencentes à OPEP em 1975 e colocando uma reta de tendência
plt.scatter(PIB_PP_1997.loc[PIB_PP_1997.OPEP_petroleo == "Não OPEP","1997_petroleo"], PIB_PP_1997.loc[PIB_PP_1997.OPEP_PIB == "Não OPEP", "1997_PIB"], color="indigo", label = "Não OPEP")
reta(PIB_PP_1997.loc[PIB_PP_1997.OPEP_petroleo == "Não OPEP","1997_petroleo"], PIB_PP_1997.loc[PIB_PP_1997.OPEP_PIB == "Não OPEP", "1997_PIB"], "indigo")


plt.grid()
plt.title('Produção de petróleo x Crescimento anual do PIB em 1997')
plt.ylabel('Crescimento anual do PIB')
plt.xlabel('Produção de petróleo')
plt.legend()
```

```python
#Plotando o gráfico de produção de petróleo anual x crescimento do PIB 
plt.scatter(PIB_PP_2013.loc[PIB_PP_2013.OPEP_petroleo == "OPEP","2013_petroleo"], PIB_PP_2013.loc[PIB_PP_2013.OPEP_PIB == "OPEP", "2013_PIB"], color="turquoise", label = "Países OPEP")
reta(PIB_PP_2013.loc[PIB_PP_2013.OPEP_petroleo == "OPEP","2013_petroleo"], PIB_PP_2013.loc[PIB_PP_2013.OPEP_PIB == "OPEP", "2013_PIB"], "turquoise")

plt.scatter(PIB_PP_2013.loc[PIB_PP_2013.OPEP_petroleo == "Não OPEP","2013_petroleo"], PIB_PP_2013.loc[PIB_PP_2013.OPEP_PIB == "Não OPEP", "2013_PIB"], color="indigo", label = "Não OPEP")
reta(PIB_PP_2013.loc[PIB_PP_2013.OPEP_petroleo == "Não OPEP","2013_petroleo"], PIB_PP_2013.loc[PIB_PP_2013.OPEP_PIB == "Não OPEP", "2013_PIB"], "indigo")

plt.grid()
#plt.xscale('log')
plt.title('Produção de petróleo x Crescimento anual do PIB em 2013')
plt.ylabel('Crescimento anual do PIB')
plt.xlabel('Produção de petróleo')
plt.legend()
```

Em 2013 o crescimento do PIB, para os países com produções de petróleo não tão exurbitantes, o crescimento anual do PIB foi inclusive menor do que os do países não pertencentes ao cartel. 

## Conclusão

Supõe-se com esse trabalho que participar ou não da OPEP não influencia na produção de petróleo de cada país (como pode ser observado no gráfico de barras de produção de petróleo ao longo do tempo); tal fator é influenciado por outras ocasiões. Assim, a vantagem de participar da OPEP estaria apenas nas ações do cartel no mercado e na política. 

Desse modo, através dos gráficos feitos especula-se que em 1975 participar da OPEP era vantajoso para os países exportadores de petróleo. Nessa época a produção anual de petróleo dos países pertencentes à organização representava a maior parte da produção mundial (como pode ser observado pelo gráfico de setor de 1975); sendo assim, as ações do cartel geravam grandes impactos ao mercado internacional, deixando, então, os países compradores de petróleo a mercê do cartel e suas ações. Portanto, os países da OPEP conseguiam render mais dinheiro com suas exportações. Como reação, o PIB dos mesmos crescia mais rapidamente do que o PIB dos outos produtores de petróleo, fenômeno que pode ser observado no gráfico de dispersão de 1975. 

Com o passar dos anos a representatividade da OPEP no mercado internacional de petróleo decresceu muito (como visto nos gráficos de setor de 1997 e 2013), o que pode ter acontecido por diversos fatores não estudados por esse projeto. Com essa regressão da representatividade dos países da OPEP os movimentos feitos pelo cartel dentro do mercado tornaram-se menos impactantes no mercado internacional (política e econômicamente). Por consequência, com o passar do tempo, o crescimento do PIB foi diminuindo até ficar praticamente igual ao crescimento do PIB dos outros produtores de petróleo (como visto no gráfico de dispersão de 1997) e chegando em 2013 a ficar menor que o crescimento do PIB dos outros países usados como comparação.
